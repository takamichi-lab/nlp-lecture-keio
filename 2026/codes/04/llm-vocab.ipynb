# LLM の語彙を見てみよう

```python
import json
import random
from huggingface_hub import hf_hub_download
```

```python
path = hf_hub_download(
    repo_id="llm-jp/llm-jp-1.3b-v1.0",
    filename="tokenizer.json"
)

# json形式の語彙を読み込む
subword_vocab = json.load(open(path, "r", encoding="utf-8"))
subword_vocab = subword_vocab["model"]["vocab"]

# 特殊トークンを除く
regular_tokens = [
    (token, idx) 
    for token, idx in subword_vocab 
    if not token.startswith("<") and not token.startswith("▁")]

# ランダムに1000トークンを表示
sampled_tokens = random.sample(regular_tokens, 1000)
for token, idx in sampled_tokens:
    print(f"{token}: {idx}")
```
