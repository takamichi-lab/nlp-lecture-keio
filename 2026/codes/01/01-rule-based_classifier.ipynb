## 規則ベースのテキスト感情分類器を作ろう．
入力されたテキストがポジティブ(1)かネガティブ(-1)かニュートラル(0)かを判定します．

```python
# 必要なモジュールのインストール
! pip install mecab-python3 unidic-lite
import MeCab
```

```python
# 特徴抽出
def extract_features(x):
    positive_words = ["最高", "素晴らしい", "良い", "好き", "楽しい", "幸せ"]
    negative_words = ["最低", "最悪", "悪い", "嫌い", "悲しい", "怖い", "難しい"]

    x_words = MeCab.Tagger("-Owakati").parse(x).split()
    features = {}
    for x_word in x_words:
        if x_word in positive_words:
            features["positive"] = features.get("positive", 0) + 1
        elif x_word in negative_words:
            features["negative"] = features.get("negative", 0) + 1

    return features
```

```python
# スコア計算
def calculate_score(features):
    feature_weights = {
        "positive": 1.0,
        "negative": -1.0
    }

    score = 0.0
    for feature, count in features.items():
        weight = feature_weights.get(feature, 0.0)
        score += weight * count
    return score
```

```python
# 決定 (classification)
def determine_sentiment(score):
    if score > 0.0:
        return "ポジティブ"
    elif score < 0.0:
        return "ネガティブ"
    else:
        return "ニュートラル"
```

```python
# 評価データを読み込み
def load_evaluation_data(file_path):
    evaluation_data = []
    with open(file_path, "r", encoding="utf-8") as f:
        for line in f:
            label, text = line.strip().split("\t")
            evaluation_data.append((text, label))
    return evaluation_data

data = load_evaluation_data("sentiment_test.txt")
print(data[:7])
```

```python
# システムを実行
for text, true_label in data:
    features = extract_features(text)
    score = calculate_score(features)
    predicted_label = determine_sentiment(score)
    t_label = {"1": "ポジティブ", "-1": "ネガティブ", "0": "ニュートラル"}[true_label]
    print(f"テキスト: {text}")
    print(f"特徴: {features}")
    print(f"スコア: {score:.2f}")
    print(f"予測: {predicted_label}, 真のラベル: {t_label}")
    print("-" * 40)
```
